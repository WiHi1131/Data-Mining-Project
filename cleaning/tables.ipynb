# Merge Profiles and Financials

Combine datasets using inner join.

```python
inner_join = pd.merge(profiles, financials, on='website', how='inner')
inner_join.head(2)
```

```python
inner_join.drop('name_y', axis=1, inplace=True)
inner_join.rename(columns={'name_x':'name'}, inplace=True)
inner_join.head(2)
```

# Merge Inner Join and Crunchbase Profiles

```python
profiles_cb = pd.read_csv('../datasets/crunchbase-profiles.csv')
```

```python
profiles_cb.head(2)
```

```python
profiles_cb.rename(columns={'homepage_url':'website'}, inplace=True)
```

```python
profiles_cb.head(2)
```

Clean up `website` column.

```python
import re

def clean_website(url):
    if isinstance(url, str):
        url = url.replace('http://', 'https://')
        url = url.replace('www.', '')
        match = re.match(r'https?://[^/]+', url)
        if match:
            return match.group()
    return None

df = pd.DataFrame()
profiles_cb['website'] = profiles_cb['website'].apply(clean_website)
profiles_cb.head(2)
```

Combine datasets using left join where previous inner join is left side.

```python
left_inner_join = pd.merge(inner_join, profiles_cb, on='website', how='left')
left_inner_join.head(2)
```

```python
left_inner_join.shape[0]
```

```python
left_inner_join.to_csv('../datasets/startups.csv', index=False) 
```
