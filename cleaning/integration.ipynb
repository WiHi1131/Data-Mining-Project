# Integration

```python
import pandas as pd

profiles = pd.read_csv('../datasets/clean-profiles.csv')
financials = pd.read_csv('../datasets/clean-financials.csv')
overviews = pd.read_csv('../datasets/clean-overviews.csv')
```

### First inner join Profiles and Financials

For all datasets, we will be joining them on the column `website`.

```python
profiles_financials = pd.merge(profiles, financials, on='website', how='inner')
profiles_financials.head(2)
```

```python
profiles_financials.drop('name_y', axis=1, inplace=True)
profiles_financials.rename(columns={'name_x':'name'}, inplace=True)
profiles_financials.head(2)
```

```python
profiles_financials.shape[0]
```

### Next inner join (Profiles $\cap$ Financials) and Overviews

```python
profiles_financials_overviews = pd.merge(profiles_financials, overviews, on='website', how='inner')
profiles_financials_overviews.head(2)
```

```python
profiles_financials_overviews.drop('name_y', axis=1, inplace=True)
profiles_financials_overviews.rename(columns={'name_x':'name'}, inplace=True)
profiles_financials_overviews.head(2)
```

```python
profiles_financials_overviews.shape[0]
```

```python
profiles_financials_overviews.info()
```

```python
profiles_financials_overviews.isnull().sum()
```

Lets drop the 8 rows missing `funding_last_date` and the 57 rows missing `year_founded`. 

```python
profiles_financials_overviews = profiles_financials_overviews.dropna(subset=['funding_last_date', 'year_founded'])
profiles_financials_overviews.isnull().sum()
```

# Rename and rearrange columns

```python
profiles_financials_overviews = profiles_financials_overviews.rename(columns={
    'cb_description': 'description',
    'valuation_numeric_millions': 'valuation_millions',
    'funding_total_numeric_millions': 'funding_total_millions',
    'funding_last_numeric_millions': 'last_funding_millions',
    'funding_last_type': 'last_funding_type',
    'funding_last_date': 'last_funding_date',
    'revenue_total_numeric_millions': 'revenue_total_millions',
    'postal': 'postal_code'
})
```

```python
profiles_financials_overviews.head(1)
```

```python
profiles_financials_overviews = profiles_financials_overviews[[
    # Info
    'name', 'tagline', 'summary', 'description',
    
    # Year founded
    'year_founded',

    # Website and location
    'website', 'city', 'region', 'country', 'postal_code',

    # Concepts and Keywords
    'concepts', 'keywords', 
    
    # Perception
    'investor_count', 'mosaic_change',
    
    # Financial
    'funding_total_millions', 'last_funding_millions', 'funding_count', 
    'last_funding_type', 'last_funding_date', 'valuation_millions', 'valuation_date',
    'revenue_total_millions', 'revenue_year',

    # Sentiment
    'sentiment', 'articles'
]]
```

```python
profiles_financials_overviews.info()
```

# Splitting

We will have three tables:
1. A primary table called `primary.csv`
2. Rows from primary table with revenue information called `primary-with-revenue.csv`
3. Rows from primary table with valuation information called `primary-with-valuation.csv`

First, we will create 2:

```python
primary_with_revenue = profiles_financials_overviews[profiles_financials_overviews['revenue_total_millions'].notnull()]
primary_with_revenue.head(2)
```

Drop valuation columns:

```python
primary_with_revenue = primary_with_revenue.copy()

primary_with_revenue.drop(['valuation_millions', 'valuation_date'], axis=1, inplace=True)
```

```python
primary_with_revenue.info()
```

Second, we will create 3:

```python
primary_with_valuation = profiles_financials_overviews[profiles_financials_overviews['valuation_millions'].notnull()]
primary_with_valuation.head(2)
```

Drop valuation columns:

```python
primary_with_valuation = primary_with_valuation.copy()

primary_with_valuation.drop(['revenue_total_millions', 'revenue_year'], axis=1, inplace=True)
```

```python
primary_with_valuation.info()
```

Finally, we will create 1:

Drop revenue and valuation columns:

```python
primary = profiles_financials_overviews.copy()

primary.drop(['revenue_total_millions', 'revenue_year', 'valuation_millions', 'valuation_date'], axis=1, inplace=True)
```

```python
primary.info()
```
